# Extracting Post URLs 

```python
# scroll using a for loop 30 times to reach around post 400 of most recent on the page 
posts = []
for i in range(5):
    ActionChains(driver).send_keys(Keys.END).perform()
    time.sleep(5)
    links = []
    links = driver.find_elements_by_tag_name("a")
    # create a loop to iterate through the href url links and add them to the posts list 
    for link in links:
        post = link.get_attribute("href")
        # to ensure that we only extract posts and not other href attributes: 
        if "/p/" in post:
            if post not in posts: posts.append(post)
print(len(posts))
```

```python
# source used: https://www.kirenz.com/post/2021-12-11-text-mining-and-sentiment-analysis-with-nltk-and-pandas-in-python/text-mining-and-sentiment-analysis-with-nltk-and-pandas-in-python/
# Text pre-processing --> I will conduct tokenization followed by removal of stopwords, stemming and perform lemmatization
from nltk.tokenize import RegexpTokenizer
# define tokanizer 
regexp = RegexpTokenizer('\w+')
#tokanize comments in dataframe 
df2['Tokenized']=df2['Demojified'].apply(regexp.tokenize)
df2.head(20)
```

```python
from nltk.corpus import stopwords 
# make list of english stopwords 
stopwords = nltk.corpus.stopwords.words("english")
# extend the list with custom stopwords
my_stopwords = ['https']
# stopwords.extend(my_stopwords) -> incase needed 
# use lambda function to remove stopwords 
df2['Token_no_stopwords'] = df2['Tokenized'].apply(lambda x: [item for item in x if item not in stopwords])
df2.head(20)
```

```python
# removing infrequent words that appear less than 2 times
df2['Token_no_infrequent'] = df2['Token_no_stopwords'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))
df2.head(20)
```

```python
# Needed for lemmatizer 
nltk.download('omw-1.4')
nltk.download('wordnet')
```

```python
# perform lemmatization 
from nltk.stem import WordNetLemmatizer 
wordnet_lem = WordNetLemmatizer()
df2['Lemmatized'] = df2['Token_no_infrequent'].apply(wordnet_lem.lemmatize)
df2
```
